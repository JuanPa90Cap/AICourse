## AnÃ¡lisis inicial


*   Tenemos 412698 filas y 53 columnas
*   Hay muchos valores faltantes
*   La columna saledate es de tipo objeto, tenemos que convertir en un objeto datetime.
*   La mayorÃ­a de las columnas son de tipo de dato string, le podemos convertir en un tipo de dato categÃ³rico





```python
df1 = pd.read_csv("/content/drive/MyDrive/stock/TrainAndValid.csv", low_memory=False, parse_dates=['saledate'])
```

```python
df1.saledate.head(5)
```

Ahora tenemos la columna saledate como "datetime" pero vemos que estÃ¡ desorden por lo que vamos a ordendar de forma ascendente.

### Ordenar el Dataframe por saledate

```python
df1.sort_values(by=["saledate"], inplace=True, ascending=True)
```

```python
df1.saledate.head(3)
```

## Creamos nuevas columnas
Creamos las columnas aÃ±o(saleyear),mes(salemonth), dÃ­a(saleday), dÃ­a de la semana(saledayofweek), dÃ­a del aÃ±o(saledayofyear)

```python
df1["saleyear"] = df1.saledate.dt.year
df1["salemonth"] = df1.saledate.dt.month
df1["saleday"] = df1.saledate.dt.day
df1["saledayofweek"] = df1.saledate.dt.dayofweek
df1["saledayofyear"] = df1.saledate.dt.dayofyear

df1.head(3)
```

```python
df1.drop("saledate", axis=1, inplace=True)
```

### Convertir los string en category
Strings no se utiliza en los modelos de Machine Learning por lo que necesitamos convertirlo en category y luego en tipo de dato numÃ©rico

```python
list(df1.dtypes[df1.dtypes == 'object'].index)
```

```python
string_columns = df1.dtypes[df1.dtypes == 'object'].index
string_columns
```

```python
for col in string_columns:
  df1[col] = df1[col].astype('category').cat.as_ordered()
df1.info()
```

#### Otra forma de iterar los elementos de un DataFrame

```python
#No correr esta linea de cÃ³digo
for label, content in df1.items():
    if pd.api.types.is_string_dtype(content):
        df1[label] = content.astype("category").cat.as_ordered()

df1.info()
```

### Completar valores faltantes de columnas con tipo de dato numÃ©rico

```python
#Cuales columnas con tipo numÃ©rica
for label, content in df1.items():
  if pd.api.types.is_numeric_dtype(content):
    print(label)
```

```python
#Cuales son las columnas con datos numÃ©ricos faltantes
for label, content in df1.items():
  if pd.api.types.is_numeric_dtype(content):
    if pd.isnull(content).sum():
      print(label)
```

```python
for label, content in df1.items():
  if pd.api.types.is_numeric_dtype(content):
    if pd.isnull(content).sum():
      df1[label] = content.fillna(content.median())
```

```python
#Comprubo que en las dos columnas ya no tenga datos faltantes
df1['auctioneerID'].isnull().sum()
```

```python
#Comprubo que en las dos columnas ya no tenga datos faltantes
df1['MachineHoursCurrentMeter'].isnull().sum()
```

## Completar los valores faltantes(CategÃ³ricos)

```python
#Columnas con valores faltantes
for label, content in df1.items():
  if not pd.api.types.is_numeric_dtype(content):
    if pd.isnull(content).sum():
      print(label)
```

```python
for label, content in df1.items():
  if not pd.api.types.is_numeric_dtype(content):
    df1[label+"_is_missing"]  = pd.isnull(content)
    df1[label] = pd.Categorical(content).codes+1
```
